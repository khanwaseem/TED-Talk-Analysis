```python
# Convert the datatypes to int8 once for all column and igone the errors occurred in string type columns
mydata = mydata.astype(dtype = "int32",errors = "ignore")
```

```python
# Delete non usefull column from the dataset
mydata.drop("film_date",axis=1,inplace = True)
```

```python
mydata["description"]= mydata["description"].str.lower().str.strip()
mydata["event"]= mydata["event"].str.lower().str.strip()
mydata["name"]= mydata["name"].str.lower().str.strip()
mydata["title"]= mydata["title"].str.lower().str.strip()
mydata['published_date'] = pd.to_datetime(mydata['published_date'], unit='s')
```

```python
mydata.head(2)
```

```python
mydata.ratings = mydata.ratings.apply(ast.literal_eval)
```

```python
mydata.ratings[0]
```

```python
# Check the number of null value in Each column
mydata.isna().sum()
```

```python
mydata['speaker_occupation'].fillna(value="Not_Available", inplace=True)
```

```python
mydata.isna().sum()
```

```python
#Add two new columns
for i, rate in enumerate(mydata.ratings):
    maximum = 0
    max_name = ""
    cnt_rating = 0
    funny_count = 0
    for item in rate:
        name = item["name"]
        count = item["count"]
        if name=="Funny":
            funny_count = count
        cnt_rating+=count
        if count>maximum:
            maximum=count
            max_name = name
    mydata.loc[i,"max-rated"] = max_name
    mydata.loc[i,"max-count"] = cnt_rating
    mydata.loc[i,"funny"] = funny_count
    
```
